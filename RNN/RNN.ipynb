```python
#importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#read the dataset
dataset=pd.read_csv("Google_Stock_Price_Train.csv")
dataset
```

```python
#take the necessary columns
train_set=dataset.iloc[:,1:2].values
```

```python
train_set
```

```python
#apply scaling
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
Scaled_training=sc.fit_transform(train_set)
```

```python
Scaled_training
```

```python
#Split the data into input and output
x_train=[]
y_train=[]
for i in range(60,1258):
    x_train.append(Scaled_training[i-60:i,0])
    y_train.append(Scaled_training[i,0])
```

```python

x_train
```

```python
x_train,y_train=np.array(x_train),np.array(y_train)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_train.ndim
```

```python
#LSTM only accept three dimensional arrays
x_train=np.reshape(x_train,(1198,60,1))
```

```python
x_train.ndim
```

```python
#model building
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
model=Sequential()
```

```python
model.add(LSTM(units=50,input_shape=(x_train.shape[1],1),return_sequences=True))
```

```python
model.add(LSTM(units=50,return_sequences=True))
```

```python
model.add(LSTM(units=50,return_sequences=True))
```

```python
model.add(LSTM(units=50))
```

```python
model.add(Dense(units=1))
```

```python
model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x_train, y_train,epochs=5,batch_size=32)
```

```python
#Test the model
dataset_test=pd.read_csv("Google_Stock_Price_Test.csv")
dataset_test
```

```python
#Take the necessary columns
real_stock_price=dataset_test.iloc[:,1:2].values
real_stock_price
```

```python
dataset_total=pd.concat((dataset['Open'],dataset_test['Open']),axis=0)
dataset_total
```

```python
dataset_total.shape
```

```python
inputs=dataset_total[len(dataset_total)-len(dataset_test)-60:].values
```

```python
inputs.shape
```

```python
inputs=inputs.reshape(-1,1)
```

```python
inputs=sc.transform(inputs)
```

```python
x_test=[]
for i in range(60,80):
    x_test.append(inputs[i-60:i,0])
```

```python
x_test=np.array(x_test)
x_test.shape
```

```python
x_test=np.reshape(x_test,(20,60,1))
```

```python
predicted_stock_price=model.predict(x_test)
predicted_stock_price
```

```python
sc.inverse_transform(predicted_stock_price)
```

```python
#plot a graph between y_test and predicted y_test
plt.plot(real_stock_price,color="red",label="real_stock_price")
plt.plot(sc.inverse_transform(predicted_stock_price),color="blue",label="predicted_stock_price")
```
